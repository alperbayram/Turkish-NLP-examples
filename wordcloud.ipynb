```python
from pandas import Series, DataFrame
df = pd.read_csv(data_path + 'TrkceTwit.csv')
df_Tweets = pd.DataFrame(df)
```

```python
text = df_Tweets.Tweets[4000]
wordcloud = WordCloud().generate(text)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
wordcloud = WordCloud(max_font_size=50, max_words=100, background_color="white").generate(text)
plt.figure()
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```

```python
text = " ".join(review for review in df_Tweets.Tweets)
print ("There are {} words in the combination of all review.".format(len(text)))
```

```python
stopwords = set(STOPWORDS)
stopwords.update(["ben", "sen", "biz", "de", "ki", "ve", "da", "bir", "ile", "için", "bu", "ne", "gibi"])

wordcloud = WordCloud(stopwords=stopwords, background_color="white").generate(text)
plt.figure(figsize=[10,10])
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")

plt.show()
```

  **Burada tüm tweetleri tek bir metin haline getirdik

```python
total_items_orders = df_Tweets.Tweets.sum()
total_items_orders
```

```python
stopwords = set(STOPWORDS)
stopwords.update(["ben", "sen", "biz", "de", "ki", "ve", "da", "bir", "ile", "için", "bu", "ne", "gibi"])

wordcloud = WordCloud(stopwords=stopwords, background_color="white").generate(total_items_orders)
plt.figure(figsize=[10,10])
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")

plt.show()
```
